```python
import json
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem.lancaster import LancasterStemmer
import numpy as np
import random
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Layer,Input,Dense
import tensorflow
import pickle
```

```python
from tensorflow.python.framework import ops
ops.reset_default_graph()
```

```python
stemmer=LancasterStemmer()
```

```python
data=pd.read_json('intents.json')
```

```python
#data.loc[1]
```

```python
try:
    with open("data.pickle","rb") as f:
        bags,labels,training,output=pickle.load(f)
except:
    bags=[]
    labels=[]
    dock_x=[]
    dock_y=[]

    for intents in data.intents:
        for pattern in intents['patterns']:
            w=word_tokenize(pattern)
            bags.extend(w)
            dock_x.append(pattern)
            dock_y.append(intents['tag'])

        if intents['tag'] not in labels:
            labels.append(intents['tag'])

    bags=[stemmer.stem(w) for w in bags]
    bags=sorted(list(set(bags)))

    labels=sorted(labels)

    training=[]
    output=[]

    out_empty=[0 for _ in range(len(labels))]

    for x,doc in enumerate(dock_x):
            bag=[]

            wrds=[stemmer.stem(w) for w in doc]
            for w in bags:
                if w in wrds:
                    bag.append(1)
                else:
                    bag.append(0)

            out_row=out_empty[:]
            out_row[labels.index(dock_y[x])]=1

            training.append(bag)
            output.append(out_row)

    training=np.array(training)
    output=np.array(output)
    
    with open("data.pickle","wb") as f:
        pickle.dump((bags,labels,training,output),f)
```

```python
in_layer=Input(shape=training.shape[1])
d1=Dense(8,activation="sigmoid")(in_layer)
d2=Dense(8,activation='sigmoid')(d1)
out=Dense(output.shape[1],activation="softmax")(d2)
```

```python
model=Model(inputs=in_layer,outputs=out)
```

```python
model.compile(optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
try:
    model.load("model.tflearn")
except:
    outs=model.fit(training,output,epochs=1000)
    model.save("model.tflearn")
```

```python
def bag_of_word(s,bag):
    bag_z=[0 for _ in range(len(bag))]
    
    s_words=word_tokenize(s)
    s_word=[stemmer.stem(w.lower()) for w in s_words]
    
    for w in s_word:
        for i,wo in enumerate(bag):
            if w==wo:
                bag_z[i]=1
               
    return np.array(bag_z)
```

```python
def chat():
    print("Start talking with the bot or type quit to stop talking")
    while True:
        inp=input("input:")
        if inp.lower()=='quit':
            break
        
        #res=[bag_of_word(inp,bags)]
        #res_np=np.array(res)
        #print(res_np.shape)
        result=model.predict(np.array([bag_of_word(inp,bags)]))
        result_index=np.argmax(result)
        tag=labels[result_index]
    
        for tg in data['intents']:
            if tg['tag']==tag:
                response=tg['responses']
        print(random.choice(response))
```

```python
chat()
```
